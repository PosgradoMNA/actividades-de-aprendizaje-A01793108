## 1.2 Limpieza de datos

Vamos a confirmar los prosibles errores o ruidos en algunas categorías

```python
print(nData.groupby(['Genero']).size()) ## no existen novedades
print(nData.groupby(['EstadoCivil']).size()) ## el estado civil tienen 4 niveles cuando se menciona 3 
print(nData.groupby(['NivelEducacion']).size()) ### al parecer hay 6 categorias cuando debe haber 4
#print(nData.groupby(['Edad']).size())
```

```python
modaEstadoCivil = nData['EstadoCivil'].mode()[0]
```

```python
nData['EstadoCivil'] = nData['EstadoCivil'].map({0:modaEstadoCivil, 1:1, 2:2, 3:3}) 
nData['NivelEducacion'] = nData['NivelEducacion'].map({0:4, 1:1, 2:2, 3:3, 4:4, 5:4, 6:4}) 
```

**En cuanto a los valores numéricos**, para encontrar alguna inconsistencia, tendríamos que conocer un poco más sobre la naturaleza de los mismos, para identificar cuando existen valores poco lógicos.

Sin embargo, podemos ver una alta dispeción en algunas, el valor máximo se distancia notablemente del tercer cuartil.
