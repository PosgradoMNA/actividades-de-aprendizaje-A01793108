## 2.2 Dealing with missingg values

los valores perdidos puedne representarse con: "?", "N/A", 0 or blank cell

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df = pd.read_csv(filename, names = headers)
df.replace("?", np.nan, inplace = True)
df.head(10)
```

```python
print(df.shape)
print(df.dropna().shape) ## especificar 0 rows o 1 columns
print(df.dropna(axis = 0).shape)
print(df.dropna(axis = 1).shape)
print(df.dropna(subset = ["price"], axis = 0).shape)
```

```python
df.isnull().any()
```

```python
## remplazar na con valores
mean = (pd.to_numeric(df["normalized-losses"])).mean()
print(mean)
df["normalized-losses"].replace(np.nan, mean, inplace = True)
```

## 2.3 Data formatting

al recolectar informacion de distintas fuentes pueden estar en distitos formatos y abreviaturas, por lo que se tiene que darle el mismo formato

Facilita el análisis al homologar los grupos

```python
df["city-mpg"] = 235/df["city-mpg"] 
df.rename(columns = {"city-mpg": "city-L/100km"}, inplace = True)
df.head()
```

```python
df.info()
```

```python
df["normalized-losses"] = df["normalized-losses"].astype("int") ## no puedo transformar en int NaN
```

## 2.4 Data Normalization

La normalizacion es una etapa enscencial para el entrenamiento de algoritmos.

Ayuda a que la escala de las distintas variables no afecten al modelo de forma desigual solo por su unidad de medida, sino por que en realidad son relevantes.

```python
print(min(df["normalized-losses"]))
print(max(df["normalized-losses"]))
print(np.std(df["normalized-losses"]))
print(np.mean(df["normalized-losses"]))
```

```python
# simple feature scaling 
df["normalized-losses"]/np.max(df["normalized-losses"])
```

```python
print(np.max(df["normalized-losses"]/np.max(df["normalized-losses"])))
print(np.min(df["normalized-losses"]/np.max(df["normalized-losses"])))
print(np.std(df["normalized-losses"]/np.max(df["normalized-losses"])))
print(np.mean(df["normalized-losses"]/np.max(df["normalized-losses"])))
```

```python
# min - max
(df["normalized-losses"]-np.min(df["normalized-losses"]))/(np.max(df["normalized-losses"])-np.min(df["normalized-losses"])) 
```

```python
print(np.max( (df["normalized-losses"]-np.min(df["normalized-losses"]))/(np.max(df["normalized-losses"])-np.min(df["normalized-losses"])) ))
print(np.min( (df["normalized-losses"]-np.min(df["normalized-losses"]))/(np.max(df["normalized-losses"])-np.min(df["normalized-losses"])) ))
print(np.std( (df["normalized-losses"]-np.min(df["normalized-losses"]))/(np.max(df["normalized-losses"])-np.min(df["normalized-losses"])) ))
print(np.mean( (df["normalized-losses"]-np.min(df["normalized-losses"]))/(np.max(df["normalized-losses"])-np.min(df["normalized-losses"])) ))
```

```python
# z-score 
(df["normalized-losses"]-np.mean(df["normalized-losses"]))/np.std(df["normalized-losses"])
```

```python
print(np.max( (df["normalized-losses"]-np.mean(df["normalized-losses"]))/np.std(df["normalized-losses"]) ))
print(np.min( (df["normalized-losses"]-np.mean(df["normalized-losses"]))/np.std(df["normalized-losses"]) ))
print(np.std( (df["normalized-losses"]-np.mean(df["normalized-losses"]))/np.std(df["normalized-losses"]) ))
print(np.mean( (df["normalized-losses"]-np.mean(df["normalized-losses"]))/np.std(df["normalized-losses"]) ))
```

## 2.5 Binning

```python
df["horsepower"].replace(np.nan, mean, inplace = True)
df["horsepower"] = df["horsepower"].astype(int, copy = True)
plt.hist(df["horsepower"])
```

```python
bins = np.linspace(df["horsepower"].min(), df["horsepower"].max(), 4)
bins
```

```python
groups_names = ["low", "medium", "high"]
### generar bins
# df["horsepower-binned"] = 
pd.cut(df["horsepower"], bins, labels = groups_names, include_lowest = True) ## mantiene orden 
```

```python
df["horsepower-binned"] = pd.cut(df["horsepower"], bins, labels = groups_names, include_lowest = True) ## mantiene orden 
plt.bar(groups_names, df["horsepower-binned"].value_counts())
```

```python
df["horsepower-binned"] = pd.cut(df["horsepower"], bins, labels = groups_names, include_lowest = True)
df[["horsepower", "horsepower-binned"]].head(20)
```
