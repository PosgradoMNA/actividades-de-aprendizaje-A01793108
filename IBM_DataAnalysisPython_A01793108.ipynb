## 3.3 ANOVA
**A**nalysis **O**f **VA**riance 

permite encontrar la correlacion que tienen las variables categoricas.

para eso se utiliza la prueba de hipotesis f para comparar las medias de los diferentes grupos y la prueba t para saber si esta es significativa.

```python
df_grp = df[['make','price']].groupby('make', as_index = False).mean()
order = df.groupby('make')['price'].mean().sort_values().index
sns.barplot(x = 'price', y = 'make', data = df_grp, order = order)
```

```python
df_anova = df[['make','price']].groupby('make')
df_anova.get_group('audi')['price']
```

```python
f_val, p_val = stats.f_oneway(df_anova.get_group('audi')['price'], df_anova.get_group('bmw')['price'])  
print( "ANOVA results: F=", f_val, ", P =", p_val)   
```

```python
f_val, p_val = stats.f_oneway(df_anova.get_group('peugot')['price'], df_anova.get_group('saab')['price'])  
print( "ANOVA results: F=", f_val, ", P =", p_val)   
```

## 4.1 Modelo lineal simple y multiple

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import Pipeline
```

```python
filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df = pd.read_csv(filename, names = headers)
df.replace("?", np.nan, inplace = True)
df.head(10)
```

```python
df.info()
```

```python
df['price'].replace(np.nan, inplace = True)
df['price'] = df['price'].astype(int)
## simple linear regression
X = df[['highway-mpg']]
y = df['price']
```

```python
lmS = LinearRegression()
lmS.fit(X, y) 
yhatS = lmS.predict(X) 
```

```python
print(lmS.intercept_)
print(lmS.coef_)
```

```python
ax1 = sns.distplot(df['price'], hist = False, color = 'r', label = 'Actual Value')
sns.distplot(yhatS, hist = False, color = 'b', label = 'Fitted Values', ax = ax1)
```

```python
## multiple linear regression
df['price'].replace(np.nan, inplace = True)
df['price'] = df['price'].astype(int)
df['horsepower'].replace(np.nan, inplace = True)
df['horsepower'] = df['horsepower'].astype(int)
```

```python
X = df[['highway-mpg', 'engine-size', 'curb-weight', 'horsepower']]
y = df['price']
```

```python
lmM = LinearRegression()
lmM.fit(X, y) 
yhatM = lmM.predict(X) 
```

```python
print(lmM.intercept_)
print(lmM.coef_)
```
